```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn import cluster, datasets
from sklearn.cluster import KMeans
```

```python
# X, y = datasets.make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)
# X, y = datasets.make_moons(200, noise=.03, random_state=0)
# X, y = datasets.make_moons(200, noise=.03, random_state=0)
# X, y = datasets.make_circles(n_samples=300, factor=0.5, noise=0.02)
X, y =np.random.rand(200,2), None


plt.scatter(X[:,0], X[:,1])
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=2, init='k-means++', max_iter=300, n_init=10, random_state=0)
pred_y = kmeans.fit_predict(X)
plt.scatter(X[:,0], X[:,1],c=pred_y)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red')
plt.show()
```
